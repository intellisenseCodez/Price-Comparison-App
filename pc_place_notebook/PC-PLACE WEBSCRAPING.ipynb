# Import the neccesary libraies

```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

# Fetch the data

```python
product_data = {
    "prod_title": [],
    "current_price": [],
    "old_price": [],
    "discount": [],
    "rating": [],
    "photo": [],
    "URL": [],
    "vendor": []
}

for page_num in range(1, 31):
    URL = f"https://pcplaceng.com/product-category/all-laptops/page/{page_num}"
     
    try:
        response=requests.get(url=URL)
        if response.status_code == 200:
            content = response.content
        else:
            print("resource not found")
    except:
        pass
    
    soup = BeautifulSoup(content, "html.parser")
    
    #find prod_details
    prod_details = soup.find_all("div", class_="mf-product-details-hover")
    prod_thumbnails = soup.find_all("div", class_="mf-product-thumbnail")
    
    # looping the articles
    for prod_detail in prod_details:
        
        product_data['discount'].append("")
        product_data['rating'].append("")
        product_data['vendor'].append("PC-PLACE_site") 
        
        prod_title = prod_detail.find("h2", class_="woo-loop-product__title")
        if prod_title != None:
            product_data['prod_title'].append(prod_title.text)
        else:
            product_data['prod_title'].append("")
            
        current_price = prod_detail.find("bdi")
        if current_price != None:
            product_data['current_price'].append(current_price.text)
        else:
            product_data['current_price'].append("")
            
        old_price = prod_detail.find("del")
        if old_price != None:
            product_data['old_price'].append(old_price.text)
        else:
            product_data['old_price'].append("")  
            
    
    for thumbnail in prod_thumbnails:
        photo = thumbnail.find("img").get('src')
        if photo != None:
            product_data['photo'].append(photo)
        else:
            product_data['photo'].append("")
            
                  
        prod_url = thumbnail.find("a").get('href')
        if URL != None:
            product_data['URL'].append(prod_url)
        else:
            product_data['URL'].append("")
            
    print(f"Done Collecting Data from {URL}")
```

# Store into pandas

```python
#store into dataframe
pcplace_laptop_df = pd.DataFrame.from_dict(product_data)
pcplace_laptop_df
```

```python
#convert the old and current price to string to remove the signs behind
pcplace_laptop_df['current_price']=pcplace_laptop_df['current_price'].astype(str)
pcplace_laptop_df['old_price']=pcplace_laptop_df['old_price'].astype(str)
```

```python
#Removing the naira sign and comma
pcplace_laptop_df['current_price']=pcplace_laptop_df['current_price'].str.replace('₦', '').str.replace(',', '')
pcplace_laptop_df['old_price']=pcplace_laptop_df['old_price'].str.replace('₦', '').str.replace(',', '')
```

```python
#convert back to numeric
pcplace_laptop_df['current_price']=pd.to_numeric(pcplace_laptop_df['current_price'])
pcplace_laptop_df['old_price']=pd.to_numeric(pcplace_laptop_df['old_price'])
```

```python
pcplace_laptop_df.loc[:,['current_price', 'old_price']] = pcplace_laptop_df[['current_price', 'old_price']].fillna(0)
```
