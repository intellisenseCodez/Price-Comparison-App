### Step 1: Importing neccessary libaries

```python
# importing our necessary libaries
import requests # import our request libary
from bs4 import BeautifulSoup # importing beautifulsoup
import pandas as pd  # importing pandas
```

### Step 2: Store Data Into Dictionary

```python
# Storing all the data into a dictionary
# Getting the information for the whole pages
product_data = {
    "Product Name": [],
    "Current Price": [],
    "Old Price": [],
    "Discount": [],
    "Ratings": [],
    "URL": [],
    "Photo": [],
    "Vendor": []
}
```

### Step 3: Working on multiple pages

```python

# Writting a for loop to get the informattion on each info needed
for page_num in range(1, 8):
    URL = f"https://www.amaget.com/product-category/computing/laptop/page/{page_num}"
    
    try:
        response = requests.get(URL)
        if response.status_code == 200:
            content = response.content
        else:
            print("Resource Not Found")
    except:
        pass
    
    soup = BeautifulSoup(content, "html.parser")
    
    # Getting the tag of a particular product and class in order to access other product also
    
    product_infos = soup.find_all('div', class_='product-inner product-item__inner')
    
    #looping the articles
    for info in product_infos:
        # product name
        name = info.find('h2', class_='woocommerce-loop-product__title')
        if name != None or '':
            product_data["Product Name"].append(name.text)
        else:
            product_data["Product Name"].append("")
        
        # current price
        current_price = info.find('span', class_="electro-price").find('ins')
        if current_price != None or '':
            product_data["Current Price"].append(current_price.text)
        else:
            product_data["Current Price"].append("")
        
        # old price
        old_price = info.find('span', class_="electro-price").find('bdi')
        if old_price != None or '':
            product_data["Old Price"].append(old_price.text)
        else:
            product_data["Old Price"].append("")
            
        # url
        prod_url = info.find('a', class_="woocommerce-LoopProduct-link woocommerce-loop-product__link").get('href')
        if prod_url != None or '':
            product_data["URL"].append(prod_url)
        else:
            product_data["URL"].append("")
            
        # photo
        prod_photo = info.find('div', class_="product-thumbnail product-item__thumbnail").find("img").get("data-lazy-src")
        if prod_photo != None or '':
            product_data["Photo"].append(prod_photo)
        else:
            product_data["Photo"].append("")
        
        # discount
        discount = None
        product_data["Discount"].append(discount)
        
        # rating
        rating = None
        product_data["Ratings"].append(rating) 
        
        product_data["Vendor"].append("Amaget Site")
        
            
    print(f"Done Collecting Data From {URL}")
        
        
```

### Step 4: Storage of the whole data in Dataframe

```python
# store in dataframe
amaget_laptop_df = pd.DataFrame.from_dict(product_data)
amaget_laptop_df    
```

### Step 5: Storage of data into a csv file

```python
amaget_laptop_df.to_csv("amaget_laptop.csv", index = False)
```

### Step 6: Performing a data cleaning on amaget data

```python
# Reading the data on the csv in other to access and clean it
amaget_df = pd.read_csv("amaget_laptop.csv")
amaget_df
```

```python
# Removing the naira sign from the current price
amaget_df['Current Price'] = amaget_df['Current Price'].str.replace("₦", "")
```

```python
# Removing the comma from thge current price
amaget_df['Current Price'] = amaget_df['Current Price'].str.replace(",", "")
```

```python
# Changing the current price to a float value
amaget_df['Current Price'] = amaget_df['Current Price'].astype("float64")
```
